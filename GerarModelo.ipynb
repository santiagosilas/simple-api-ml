```python
from sklearn import datasets
iris = datasets.load_iris()
X, y = iris.data, iris.target
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
modelo = knn.fit(X, y)
```

```python
import joblib
joblib.dump(modelo, 'model.joblib')
```
